```python
import pandas as pd
from pathlib import Path

maybe_truth = Path("../../data/dcase/test/fibonacci_rfpsen_test1.csv")
pred = Path("../../data/dcase/test/fibonacci_cm_test1.csv")

KEY = "audio_path"
LABEL = "class_labels"

def read_space_separated_labels(path: Path, *, key=KEY, label=LABEL) -> pd.DataFrame:
    df = pd.read_csv(
        path,
        header=None,
        names=[key, label],
        sep=r"	",
        engine="python",
        dtype=str,
    )
    df[key] = df[key].str.strip()
    df[label] = df[label].str.strip()
    return df

df_truth = read_space_separated_labels(maybe_truth).rename(columns={LABEL: "y_maybe_truth"})
df_pred  = read_space_separated_labels(pred).rename(columns={LABEL: "y_pred"})

m = df_truth.merge(df_pred, on=KEY, how="inner")

print("rows truth:", len(df_truth), "rows pred:", len(df_pred), "rows merged:", len(m))
print("agreement rate:", (m["y_maybe_truth"] == m["y_pred"]).mean())
print(m.head())
```
